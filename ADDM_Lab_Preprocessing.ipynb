```python
from sklearn.preprocessing import MinMaxScaler
sl = np.array(dtf['Age']).reshape(-1,1)
scl = MinMaxScaler()
scl.fit(sl)
sl_scaled = scl.transform(sl)
dtf['Age'] = sl_scaled.reshape(1,-1)[0]
```

```python
w = pd.get_dummies(dtf.Survived,drop_first=False, prefix="Class")
```

```python
dtf.drop('Survived',inplace=True,axis=1)
```

```python
dtf = pd.concat([dtf,w],axis=1)
```
