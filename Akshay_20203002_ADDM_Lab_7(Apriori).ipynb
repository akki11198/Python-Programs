```python
#importing the libraries
import pandas as pd
from apyori import apriori
import numpy as np
```

```python
#loading the dataset
dtf = pd.read_csv('E:\Python Programs\Groceries.csv')
```

```python
dtf
```

```python
#replacing NaN values with 0
dtf.fillna(0,inplace=True)
```

```python
dtf
```

```python
#dropping a non necessery column
dtf.drop(labels="Item(s)",inplace=True,axis=1)
```

```python
dtf
```

```python
#converting the dataset to a string
rec = []
for i in range (0,9835):
    rec.append([str(dtf.values[i,j]) for j in range (0,32) if str(dtf.values[i,j]) != '0'])
```

```python
rec[1]
```

```python
rec[556]
```

```python
#applying association rule
asoc_rules = apriori(rec, min_support = 0.0065, min_confidence = 0.35, min_lift = 2.75, min_length = 3)
```

```python
asoc_results = list(asoc_rules)
```

```python
#the most frequent items as per the association rules
for item in asoc_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
